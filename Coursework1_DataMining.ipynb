## 3.  Data cleaning
- Perform data cleaning to replace any NaN values with the mean value of that particular
feature.
- Then replace any categorical features with numerical features.
- Display the resulting dataset.

```python
income= pd.read_csv('country-income.csv')
display(income.head(10))
```

```python
# Data cleaning: Missing values

print('Number of missing values:')
for col in income.columns:
    print('\t%s: %d' % (col,income[col].isna().sum())) 
```

```python
## Replace missing values with feature's mean

# Create list with variables containing missing variables
missing_vars=['Age', 'Income']

# Define function to fill missing values with variable's mean
def fill_missing_values(income, missing_vars):
    for var in missing_vars:
        income2= pd.to_numeric(income[var])
        income2= income2.fillna(income2.mean())
        income[var]= income2
    return income

#Apply function to dataset
income= fill_missing_values(income, missing_vars)

print('\nAfter replacing missing values:')
display(income.head(10))

#Check that there are no variables with missing values
for col in income.columns:
    print('\t%s: %d' % (col,income[col].isna().sum()))
```

```python
# Data cleaning: Replace categorical figures with numerical features

#First, encode categorical features as their correct data type
income= income.astype({'Region': 'category', 'Age':'float64', 'Income':'float64', 'Online Shopper': 'category'})

# Use sklearn.preprocessing package to encode categorical variables as numerical variables
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder

categorical_columns=['Region','Online Shopper' ]
def categorical_encoder(dataset, categorical_columns):
    for column in categorical_columns:
        encoder= LabelEncoder()
        encoded_labels= encoder.fit_transform(income[column])
        income[column]= encoded_labels
    return dataset
    
income= categorical_encoder(income, categorical_columns)

```

## 5.  Using the wine dataset from question 1, perform Principal Component Analysis (PCA) with 2 components, transform the data and plot the scatterplot of all samples along the two principal components, color-coded according to the "target" column.

```python
from sklearn.datasets import load_wine

#Load wine data
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = pd.Series(data.target) 
df_wine= df.drop(columns='target') #Create dataset for PCA excluding 'target' column

df_wine.head()
```

The new attributes created from the PCA: 1) are linear combination of the original attributes from the 'wine' dataset, 2) they are orthogonal to each other, 3) capture the maximum amount of variability in the data.

```python
import pandas as pd
from sklearn.decomposition import PCA

numComponents = 2
pca = PCA(n_components=numComponents)
pca.fit(df_wine)

projected = pca.transform(df_wine)
projected = pd.DataFrame(projected,columns=['pc1','pc2'],index=range(1,df.shape[0]+1))
projected['target'] = df['target']
```

```python
# plot the scatterplot of all samples along the two principal components, color-coded according to the "target" column 
import matplotlib.pyplot as plt

colors = {0.0:'b', 1.0:'r', 2.0:'g'} 
markerTypes = {0.0 :'+', 1.0:'x', 2.0:'o'} 

for targetType in markerTypes:
    d = projected[projected['target']==targetType]
    plt.scatter(d['pc1'],d['pc2'],c=colors[targetType],s=60,marker=markerTypes[targetType])
```

- **What insights can you obtain by viewing the scatterplot of the principal components? Do samples overlap?**

Viewing the scatterplot of the principal components, we can conclude that samples from different classes can be somewhat distinguished. Samples belonging to *target type 0*, which correspond to the blue crosses, are all in the right side of the graph. Samples belonging to *target type 1*, which correspond to the red crosses, are mostly situated on the left-hand side of the graph, although there appear to be some outliers overlapping with data points with *target type 1*. Lastly, samples belonging to *target type 2*, which correspond to the green dots, appear on the middle left area of the graph, overlapping heavily with samples with *target type 1*.

Overall, there is a lot of overlapping, specially of samples with *target type 1* and *target type 2*, so distinguishing the class to which one sample belongs to would prove hard, if we cannot see the colour-coded labels.

- **Why is overlap happening?**

One of the reasons why overlap might happen is that the features in the dataset have different measurement scales, therefore, the way that PCA creates the two new attributes for the sample is affected and results in overlapping of the data. Specifically, using PCA, the number of attributes in the dataset was reduced from 13 to two attributes. These attributes are linear combinations of the previous attributes, they are orthogonal and they capture all of the variability in the data. Hence, since the original attributes had different scales, different variabilities from features resulted in samples which appeared to be similar in measurements and therefore overlapped in the graph.

- **What can be done to the data prior to performing PCA in order to alleviate this issue?**
 
An approach that can be taken to fix the issue is to standardize the data set. That, is, to use each feature or variable's mean and standard deviation so that all of the features are on the same scale.
Consequently, the PCA will capture the true variability and perform better at creating two components with which to classify the data set by.


```python
# Normalise values using z-scores
Z = (df_wine-df_wine.mean())/df_wine.std()
display(Z)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

numComponents = 2
pca = PCA(n_components=numComponents)
pca.fit(Z)

projected = pca.transform(Z)
projected = pd.DataFrame(projected,columns=['pc1','pc2'],index=range(1,df.shape[0]+1))
projected['target'] = df['target']
projected

colors = {0.0:'b', 1.0:'r', 2.0:'g'} 
markerTypes = {0.0 :'+', 1.0:'x', 2.0:'o'} 

for targetType in markerTypes:
    print(targetType)
    d = projected[projected['target']==targetType]
    plt.scatter(d['pc1'],d['pc2'],c=colors[targetType],s=60,marker=markerTypes[targetType])
```

- **Now are the different classes (quite) distinctive one from the other?**

Now, we can clearly differentiate between samples that belong to each class.
Samples on the top left corner belong to *target type 2*, samples on the bottom, middle part of the graph correspond to *target type 1* and samples on the top right corner correspond to *target type 0*.

There is a little overlapping of the red samples with both the green and blue samples, but overall the classes are very distinctive, and the PCA shows to perform well as a classifier.

## 6. Explain how the following inference from the heatmap visualisation of the distance matrix:

In Lab session 3 (Data Exploration and Data Visualisation), in subsection 1.9 you had
created and visualised a heatmap for the distance matrix for the graduation_rate.csv. You
may have noticed that the distance matrix visualisation is not very informative. However, it
is still possible to infer that the average distance between students whose parents only have
some high school education and students whose parents have a master's degree is larger
than the average distance between students whose parents only have some high school
education. Explain how this inference is possible from the visualisation.

A distance matrix consists of a table that shows the Euclidean distance between pairs of objects. To calculate these distances, features are scaled, in order for the computed distances to be comparable. 

Since the features have been sorted by 'parental level of education', they are ordered folowing the following order:

*education_order = ['some high school', 'high school', 'some college', "associate's degree", "bachelor's degree", "master's degree"]*.
 
Hence, values on the diagonal correspond to the distance between each sample and itself, with the order of values on the 'x-axis' increasing from left to right, and the order of values on the 'y-axis' increasing from top to bottom.

Therefore, by this reasoning, we obtain the following insights:
- The diagonal of the distance matrix shows a white line, because each sample is at a distance of 0 to itself.
- Samples with the lowest education level, students whose parents only have some high school education, are at a high average distance from samples with the highest education level, corresponding to students whose parents have a master's degree. We can observe this because of the darker color of the distance matrix in the bottom left corner.
This behaviour is mirrored on the top right corner, which corresponds to the average distances between the same pairs of samples, students whose parents only have some high school education and students whose parents have a master's degree.

- Similarly, on the top right corner, which corresponds to average distances between samples with low values of high school education, that is, students whose parents only have some high school education, is colour coded in a lighter blue, which implies those samples are situated at closer distances.

- Hence, comparing the observations above, from the heat map of this distance matrix, we can conclude that the average distance between students whose parents only have some high school education and students whose parents have a master's degree is larger than the average distance between students whose parents only have some high school education.

The distance matrix visualized using a heat map in Lab session 3 displays distances between pairs of samples. If the distance is higher, then the color for the data point is darker. The heat map can be visualized below.

```python
# Loading dataset (http://roycekimmons.com/tools/generated_data)
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('graduation_rate.csv')

from sklearn.preprocessing import StandardScaler

df_sorted = df.sort_values(by='parental level of education', ascending=True)
parental_education_sorted = df_sorted['parental level of education']

X = df_sorted.drop(columns='parental level of education').to_numpy()
scaler = StandardScaler()
X = scaler.fit_transform(X)
from scipy.spatial import distance

dist = distance.squareform(distance.pdist(X))
sns.heatmap(dist, square=True, xticklabels=False, yticklabels=False,
                cmap='Blues')
plt.title('Heatmap of Distance matrix of Graduation Dataset', fontsize = 10) # title with fontsize 20


plt.show()
    
plt.show()
```

**Answer:** In the following code, we load the CSV file using Cubes and create a JSON file for the data cube model.
The CSV file is created from the preprocessed income dataset. First, the dataseat is preprocessed, and then it is loaded using Cubes.



```python

# Import income file and do Preprocessing
# Data cleaning: Replace categorical figures with numerical features

## 1. Missing values: Replace missing values with feature's mean
income= pd.read_csv('country-income.csv')

missing_vars=['Age', 'Income']
def fill_missing_values(income, missing_vars):
    for var in missing_vars:
        income2= pd.to_numeric(income[var])
        income2= income2.fillna(income2.mean())
        income[var]= income2
    return income

income= fill_missing_values(income, missing_vars)


#First, encode categorical features as their correct data type
income= income.astype({'Region': 'category', 'Age':'float64', 'Income':'float64', 'Online Shopper': 'category'})

# Use sklearn.preprocessing package to encode categorical variables as numerical variables
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder

categorical_columns=['Region','Online Shopper' ]
def categorical_encoder(dataset, categorical_columns):
    for column in categorical_columns:
        encoder= LabelEncoder()
        encoded_labels= encoder.fit_transform(income[column])
        income[column]= encoded_labels
    return dataset
    
income= categorical_encoder(income, categorical_columns)
income= income.rename(columns={'Online Shopper': 'Online_Shopper'})
# saving the dataframe
income.to_csv('country-income_preprocessed.csv', index=False)

```
